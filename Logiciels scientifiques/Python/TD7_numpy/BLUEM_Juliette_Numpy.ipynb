### Estimer le nombre pi
#### Numpy

On definit les dimensions d'un domaine rectangulaire 2 m x 2 m

On definit un cercle centré dans ce rectangle avec un rayon de 1 m , ce cercle doit être contenu dans le domaine

Maintenant on tire la position (x,y) de n points dans le domaine rectangulaire

On compte le nombre de points, ndisque, contenu dans le cercle

En deduire une valeur de pi

#### Matplotlib
Tracer les points tirés aléatoirement pour un nombre de tirage relativement faible (inutile de surcharger la figure)

Tracer les points détectés à l'intérieur du cercle

Tracer le cercle (trouver la fonction matplotlib qui permet de dessiner un cercle)

#### Enregistrement
Lancez votre programme pour differentes valeurs du nombre de tirage n

Enregistrez dans un fichier .txt la valeur du nombre pi associe à la valeur n

Tracez la courbe pi = f(n)

```python
ecriture = [] # Pour vider le fichier
```

```python
# Espace de travail :
rect_absc = np.array([-1,1,1,-1,-1])
rect_ordo = np.array([-1,-1,1,1,-1])

plt.plot(rect_absc, rect_ordo)

# Cercle d'étude :
angle = np.linspace(0,2*math.pi, 40)
cercle_abs = np.cos(angle)
cercle_ordo = np.sin(angle)

plt.plot(cercle_abs, cercle_ordo)

# Determination de pi :
total = 0
incircle = 0
n = 20 # Changer le nombre de points ici
for i in range(0,n):
    total += 1
    # On prend un point random dans notre espace de travail :
    x = round(random.random()*(-1)**int(random.random()*2),2)
    y = round(random.random()*(-1)**int(random.random()*2),2)
    # On verifie si il est dans le cercle ou non
    rayon = math.sqrt(x**2 + y**2)
    if(rayon<=1):
        incircle += 1
        plt.plot(x,y, '+', color = "green")
    else :
        plt.plot(x,y, '+', color = "red")
# Calcul de pi : 
pi = round(incircle/total*4,2)
print(pi)

# Enregistrement dans un fichier :
ecriture.append(str(n) + ' , ' + str(pi) + "\n")
with open("pi_fonction_n.txt", "w") as f:
    for i in range(0, len(ecriture)):
        f.write(str(ecriture[i]))
```

Après avoir fait plusieurs essais avec plusieurs n , tracez la courbe pi = f(n) :

```python
# Courbe
essais = []
plt.axhline(3.14)
with open("pi_fonction_n.txt", "r") as f:
    for line in f:
        essais.append(line.strip().split(","))
for i in range(0, len(essais)):
    plt.plot(float(essais[i][0]), float(essais[i][1]), '.', color = "grey")
```

### Calculer la vitesse maximale et minimale atteinte par le coureur

```python
vitesseMax = coureurs['Speed (km/h)'].max()
vitesseMin = coureurs['Speed (km/h)'].min()

print("vitesse minimale :", vitesseMin, "; vitesse maximale :", vitesseMax)
```

### Créer 3 nouvelles colonnes à partir de la colonne temps: une contenant les heures, une autre les minutes et une dernière les secondes

```python
time = pandas.DataFrame(coureurs, columns = ['Time']).dropna() # On crée une table pour les temps
temps = [] # ainsi qu'une liste
for i in range(0,len(time)):
    temps.append(time._get_value(i, 'Time'))
# Dans l'emplacement [i (pour l'index); 'heures'/'minutes'/'sec'], on ajoute la donnée correspondante de notre liste temps 
for i in range(0,len(time)):
    time.at[i,'heures'] = temps[i].split(':')[0] 
    time.at[i,'minutes'] = temps[i].split(':')[1]
    time.at[i,'secondes'] = temps[i].split(':')[2]
    
print(time)
```

### Créer une nouvelle colonne de type datatime.time

```python
for i in range(0,len(time)):
    time.at[i,'DateTime'] = datetime.time(int(time.at[i,'heures']), int(time.at[i,'minutes']), int(time.at[i,'secondes']))
print(time)
```

### A l'aide de la méthode des trapèze, calculer la vitesse moyenne

```python
# Nous allons creer une fonction v(t) qui nous retourne la vitesse en fonction du temps


# Pour cela, commençons par convertir le temps hh:mm:ss en décimal :

def convTempsNb(h,m,s):
    nb = (h*60*60+m*60+s)/3600
    return nb
for i in range(0, len(time)):
    time.at[i,'temps'] = convTempsNb(int(time.at[i,'heures']),int(time.at[i,'minutes']),int(time.at[i,'secondes']))
print("table avec temps en décimal : \n", time)

# Nous ajoutons maintenant les vitesses des joueurs à cette table pour centraliser les informations (pour éviter d'écraser nos donnnées, on crée une nouvelle table):

tableTempsVitesse = time
tableTempsVitesse['Vitesse'] = coureurs['Speed (km/h)']
print("\n table centralisée : \n", tableTempsVitesse)

# On remplace les données vide par zero : (cela fausse un peu le calcul de la moyenne mais nous aurons tout de même une bonne estimation)

tableTempsVitesse = tableTempsVitesse.fillna(0)
print("\n table sans données vides: \n ",tableTempsVitesse)

# Creation de v(t) : (c'est donc un dictionnaire) 

v = {}
for i in range(0,588):
    v[round(float(tableTempsVitesse.at[i,'temps']),6)] = float(tableTempsVitesse.at[i,'Vitesse'])
print("\n notre dictionnaire v(t) : \n", v)


# Nous calculons maintenat l'aire sous la courbe de v(t) :


# Nous avons pour cela besoin de quelques variables : 

n = 588 # le nombre de données dans le tableau
a = 0 # le temps minimal
b = round(float(tableTempsVitesse.at[n,'temps']),6) # le temps maximal

aire = 0 # initialisation de l'aire
for i in range(0,n-1):
    aire += (v[round(float(tableTempsVitesse.at[i,'temps']),6)] + v[round(float(tableTempsVitesse.at[i+1,'temps']),6)])*(b/n)*1/2 # formule classique adaptée à notre stockage de données

print("\n aire :", round(aire,2))


# Nous calculons enfin la moyenne grace à l'aire sous la courbe :
moyenne = round(aire/(b-a),2)
print("\n vitesse moyenne :", moyenne)
```
