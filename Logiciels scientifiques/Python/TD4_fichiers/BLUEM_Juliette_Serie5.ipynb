## Exercice 1

Lire le fichier aeroport.txt
### Pour chaque aéroport extraire le code de l'aéroport, le nom de l'aéroport associé, la ville et le pays pour créer un dictionnaire.
La clé sera le code aéroport et la valeur contiendra le nom de l'aéroport, la ville et le pays. 

### Afficher le dictionnaire

```python
my_file = open("airport-codes.csv", "r", encoding="utf8")
dico = {}
for line in my_file:
    #print("id :", line.split(",")[0], "/ nom :",line.split(",")[2])
    table = []
    table.append(line.split(",")[2])
    table.append(line.split(",")[5])
    table.append(line.split(",")[7])
    dico[line.split(",")[0]] = table

my_file.close()

print(dico)
```

## Exercice 2

###  Ouvrir le fichier recensement.txt et compter le nombre d'habitants des 75 plus grandes villes françaises.

```python
my_recens = open("recensement.txt", "r", encoding="utf8")
lignes = []
somme = 0
for line in my_recens:
    lignes.append(line.split("\t"))
my_recens.close()

for i in range(0, int(lignes[-1][0])-1):
    somme += int(lignes[i][2])
print(somme)
```

## Exercice 3

### Ouvrir le fichier US.dat. Créer un dictionnaire Python permettant d'obtenir les synonymes d'un mot.

```python
my_synos = open("US.dat", "r", encoding="utf8")
dico = {}
nb_syno = 0
mot = ""
for line in my_synos:
    if nb_syno == 0:
        mot = line.split("|")[0]
        nb_syno = int(line.split("|")[1])
    else:
        nb_syno -= 1
        dico[mot] = line.split("|")
my_synos.close()
print(dico)
```
